```python
#bibliotecas usadas
import pandas as pd
import tweepy as tw
```

```python
#Transmitindo dados da API
bearer_token = ''
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
cliente = tw.Client(bearer_token=bearer_token, consumer_key=consumer_key, consumer_secret=consumer_secret, access_token=access_token, access_token_secret=access_token_secret)
```

```python
start = '2022-12-16T19:00:01Z'
end = '2022-12-21T16:13:01Z' #dia em que foi enviado o desafio
```

```python
num_tweets = cliente. get_recent_tweets_count( query = '@WoohooGame',start_time=start,end_time=end)
```

```python
num_tweets
```

```python
dados = num_tweets.data
```

```python
dados
```

```python
baseWoo = pd.DataFrame(dados)
```
